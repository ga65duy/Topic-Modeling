## Common vectorizer and topic model

```python
# Data 
language = 'english'
typex = 'editorial'

# Vectorization
min_df = 0.005
max_df = 0.9

# Topic Modeling
algorithm = 'lda'
num_topics = 110

train_percentage = 0.9
alpha = 'auto'
eta = 0.01
iterations = 200
passes = 60
chunksize = 5000
kappa = 0
tau_0 = 0
random_state=1
```

## Topics from wo_bigrams

```python
data_wo_bigrams = dl.get_articles_by_type(language, typex, kind="wobigrams")
texts_wo_bigrams = data_wo_bigrams['article_texts']
```

```python
vec_wo_bigrams = Vectorizer('tf', texts_wo_bigrams, min_df=min_df, max_df=max_df)
doc_token_wo_bigrams = vec_wo_bigrams.get_document_token_matrix(texts_wo_bigrams)
id2token_wo_bigrams = vec_wo_bigrams.get_id2token_mapping()
```

```python
tm_wo_bigrams = TopicModel('lda', num_topics, doc_token_wo_bigrams, id2token_wo_bigrams, alpha=alpha, eta=eta, iterations=iterations, passes=passes, chunksize=chunksize, decay=kappa, offset=tau_0, random_state=random_state)
```
