```python
import sys
sys.path.append('..')
import src.Automatic_Topic_Labeling.lable_topic as tl
import src.Automatic_Topic_Labeling.helper_functions as hp
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.topic_models import TopicModel
```

```python
german = "german"
type = "editorial"
comment = 'comment_texts'
thread = "thread_texts"
model = "nmf"
pos = None
```

```python
data_ge = dl.get_comments_by_type(german, type)
texts_editorial_g = data_ge['comment_texts']

data_gf = dl.get_forum_threads_by_language(german)
texts_forum_g = data_gf['thread_texts']

docs = texts_editorial_g +texts_forum_g
print(len(docs))
```

```python
#docs  = hp.min_word_length(docs,3)
```

```python
doc_comm = hp.load_doc(german,type,comment,"with_2bigramms")
doc_thread = hp.load_doc(german,type,thread,"with_2bigramms")
docs = doc_comm + doc_thread
print(len(docs))
```

```python
vec = Vectorizer.load('vectorizer/nmf_german_comments_combined_final.pkl')
tm = TopicModel.load('finalize_topicmodels/nmf/nmf_german_comments_combined_final_comments_125.pkl')
print(len(vec.get_id2token_mapping()))
print(tm.get_topic_token_matrix().shape)
```

```python
labels = tl.get_topic_lables([],docs,pos,model,vec,tm,n_top_words = 10, n_cand_lables = 100,lable_min_df = 5, n_labels = 8)

print("\nTopical labels:")
print("-" * 20)
for i, labels in enumerate(labels):
    print(u"Topic {}: {}\n".format(
        i,
        ', '.join(map(lambda l: ' '.join(l), labels))
    ))
```
