# ATL for english forums

```python
import sys
sys.path.append('../..')
import src.Automatic_Topic_Labeling.lable_topic as tl
import src.Automatic_Topic_Labeling.helper_functions as hp
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.topic_models import TopicModel
```

### 1-Load tags and documents

```python
data = dl.get_forum_threads_by_language("german", kind='tagged', merge_tokens=False)
thread_texts = data["thread_texts"]
thread_pos = data["thread_pos"]

pos=[('NN','NN'), ('ADJD', 'NN')]
print(len(thread_texts))
print(thread_texts[1])
print(len(thread_pos))
print(thread_pos[1])
```

### 2-Load Vectorizer and Topic model

```python
vec = Vectorizer.load('tagged/vectorizer/nmf_german_thread_texts_pos.pkl')
tm = TopicModel.load('tagged/topic_models/nmf/nmf_german_thread_texts_tagged_170.pkl')
```

### 3-Generate Topic Lables with Pos-Tags

```python
labels_pos = tl.get_topic_lables(thread_pos,thread_texts,pos,vec,tm, n_cand_lables = 100,lable_min_df = 5, n_labels = 8)
```

```python
display(hp.print_label_df(tm,labels_pos))
```

### All generated lables per topic

```python
print("\nTopical labels:")
print("-" * 20)
for i, labels in enumerate(labels_pos):
    print(u"Topic {}: {}\n".format(
        i,
        ', '.join(map(lambda l: ' '.join(l), labels))
    ))
```

## Labels without Pos-Tags

```python
labels = tl.get_topic_lables([],thread_texts,None,vec,tm,n_cand_lables = 100,lable_min_df = 5, n_labels = 5)
display(hp.print_label_df(tm,labels))
```
