### Calculate coherece and rank topics acording the coherence

### Table with coherence score and topic lables

#### Last 2 columns include topic lables

```python
writer = pd.ExcelWriter('Coherence and Topic Lables.xlsx')
```

```python
for l in(pkl_list):    
    tm = t.TopicModel.load(l)
    topics = tm.get_topics()
    topics_df = tm.get_topics_dataframe()
    new_topics = wt(topics,3)
    
    e = ExtrensicTopicLabeler()
    lables = e.get_topic_labels(topics, values=False)
    labels_preprocessed = e.get_topic_labels(new_topics, values=False)
    
    llist = []
    for ll in lables: 
        try: llist.append(ll[0])
        except Exception: 
             llist.append("-1")
        
    llpre = [] 
    for ll in labels_preprocessed:
        try: llpre.append(ll[0])
        except Exception: 
            llpre.append("-1")
        
    combined_df = e.combine_topic_and_labels(topics_df, llist)
    combined_df['preprocessed_labels'] = llpre
    
    c = tm.get_topic_coherence(per_topic= True)
    ranked = h.ranked_dataframe(combined_df,c)
    ranked.to_excel(writer,l[22:52])
    print(l)
    display(ranked)
writer.save()
    
    
```
