### Generate Topics for every single Comment

```python
# Data 
language = 'english'
typex = 'editorial'

# Vectorization
min_df = 0.005
max_df = 0.9

# Topic Modeling
algorithm = 'nmf'
#170
num_topics = 170
```

```python
data = dl.get_comments_by_type(language, typex,aggregate = "comments")
texts = data['comment_texts']
```

```python
vec = Vectorizer('tfidf', texts, min_df=min_df, max_df=max_df)
vec.save('{}_{}_{}_{}.pkl'.format(algorithm, language, typex,"per_comment"))

document_term_matrix = vec.get_document_token_matrix(texts)
id2token = vec.get_id2token_mapping()
document_term_matrix = skutil.shuffle(document_term_matrix, random_state=1)
```

```python
model = TopicModel('nmf',num_topics, document_term_matrix, id2token)
model.save('{}_{}_{}_{}_comments{}.pkl'.format(algorithm, language, typex,"per_comment",num_topics))
```
