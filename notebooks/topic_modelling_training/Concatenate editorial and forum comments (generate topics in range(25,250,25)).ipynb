#### English forum comments

```python
# Data 
english = 'english'
data = dl.get_forum_threads_by_language(english)
texts_forum = data['thread_texts']
```

#### English editorial comments

```python
editorial = 'editorial'
data = dl.get_comments_by_type(english, editorial)
texts_editorial = data['comment_texts']
```

#### Concatenate forum and editorial comments

```python
concatenated = texts_forum + texts_editorial
```

#### Initialise Vectorizer and Topic Model lda

```python
# Vectorization
min_df = 0.005
max_df = 0.9

# Topic Modeling
lda = 'lda'

train_percentage = 0.9
alpha = 'auto'
eta = 0.01
iterations = 200
passes = 60
chunksize = 5000
kappa = 0
tau_0 = 0
```

```python
vec_el = Vectorizer('tf', concatenated, min_df=min_df, max_df=max_df)
#vec_el.save('vectorizer/{}_{}_{}.pkl'.format(lda, english,"comments_combined_final"))
num_topics = 25
while num_topics < 250:
    document_term_matrix_el = vec_el.get_document_token_matrix(concatenated)
    id2token_el = vec_el.get_id2token_mapping()

    document_term_matrix_el = skutil.shuffle(document_term_matrix_el, random_state=1)
    num_docs_el, num_terms_el = document_term_matrix_el.shape
    train = int(num_docs_el * train_percentage)
    train_document_term_matrix_el = document_term_matrix_el[0:train, :]
    test_document_term_matrix_el = document_term_matrix_el[train: num_docs_el, :]

    model = TopicModel('lda', num_topics, train_document_term_matrix_el, id2token_el, random_state =1,alpha=alpha, eta=eta, iterations=iterations, passes=passes, chunksize=chunksize, test_document_term_matrix=test_document_term_matrix_el, decay=kappa, offset=tau_0)
    model.save('finalize_topicmodels/lda/{}_{}_{}_comments_{}.pkl'.format(lda, english,"comments_combined_final",num_topics))
    num_topics = num_topics + 25
```

#### English editorial and forum comments with nmf

```python
nmf = "nmf"
vec_en = Vectorizer.load("vectorizer/nmf_english_comments_combined_final.pkl")
#vec_el.save('vectorizer/{}_{}_{}.pkl'.format(nmf, english,"comments_combined_final"))

num_topics = 25
while num_topics < 250:
    document_term_matrix_en = vec_en.get_document_token_matrix(concatenated)
    id2token_en = vec_en.get_id2token_mapping()

    document_term_matrix_en = skutil.shuffle(document_term_matrix_en, random_state=1)
    num_docs_en, num_terms_en = document_term_matrix_en.shape
    train_en = int(num_docs_en * train_percentage)
    train_document_term_matrix_en = document_term_matrix_en[0:train, :]
    test_document_term_matrix_en = document_term_matrix_en[train: num_docs_en, :]

    model_n = TopicModel('nmf',num_topics, document_term_matrix_en, id2token_en,random_state =1)
    model_n.save('finalize_topicmodels/nmf/{}_{}_{}_comments_{}.pkl'.format(nmf, english, "comments_combined_final" ,num_topics))
    num_topics = num_topics +25
```
