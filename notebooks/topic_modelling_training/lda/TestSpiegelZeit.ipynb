```python
import sys
sys.path.append('../../..')
import src.Automatic_Topic_Labeling.lable_topic as tl
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.topic_models import TopicModel
import pandas as pd
import sklearn.utils as skutil
pd.set_option('display.max_rows', None)
```

```python
data = dl.get_articles_by_sources(["spiegel","zeit","tagesschau", "br"],kind = "wobigrams")
texts = data['article_texts']
```

```python
vec = Vectorizer('tf', texts, min_df=2, max_df=0.9)

document_term_matrix = vec.get_document_token_matrix(texts)
id2token = vec.get_id2token_mapping()
```

```python
document_term_matrix = skutil.shuffle(document_term_matrix, random_state=1)
num_docs, num_terms = document_term_matrix.shape
```

```python
modelt = TopicModel('lda', 15, document_term_matrix, id2token, passes=30)
```

```python
texts_splitted = [text.split() for text in texts]
labels = tl.get_topic_lables(texts_splitted, 'lda',vec,modelt,n_top_words = 15, n_cand_lables = 100,lable_min_df = 15, n_labels = 8)
```

```python
modelt.get_topics_dataframe()
```

```python
print("\nTopical labels:")
print("-" * 20)
for i, labels in enumerate(labels):
    print(u"Topic {}: {}\n".format(
        i,
        ', '.join(map(lambda l: ' '.join(l), labels))
    ))
```
