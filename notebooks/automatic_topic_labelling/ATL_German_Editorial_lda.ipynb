## ATL on German Editorial Article Texts with lda
### on tagged data

```python
import sys
sys.path.append('../..')
import src.Automatic_Topic_Labeling.lable_topic as tl
import src.Automatic_Topic_Labeling.helper_functions as hp
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.topic_models import TopicModel
```

### 1-Load tags and documents

```python
language = "german"
type = "editorial"
typex = "article"

texttype = "article_texts"
model = "lda"
pos=None

docsTuple = hp.load_taggs(language,typex)

text = hp.load_doc(language, type, texttype,"tagged")
```

### 2-Load Vectorizer and Topic model

```python
vec = Vectorizer.load('tagged/vectorizer/lda_german_editorial_pos.pkl')
tm = TopicModel.load('tagged/topic_models/lda/lda_german_editorial_tagged_articles_190.pkl')
```

### 3-Generate Topic Lables without Pos-Tags

```python
labels = tl.get_topic_lables(docsTuple,text,pos,model,vec,tm,n_top_words = 15, n_cand_lables = 100,lable_min_df = 5, n_labels = 5)
```

```python
df = tm.get_topics_dataframe()
lablelist = []
for l in labels:
    lablelist.append(' '.join(l[0]))
df["lable"] = lablelist
display(df)
```

```python
print("\nTopical labels:")
print("-" * 20)
for i, labels in enumerate(labels):
    print(u"Topic {}: {}\n".format(
        i,
        ', '.join(map(lambda l: ' '.join(l), labels))
    ))
```
