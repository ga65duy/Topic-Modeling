```python
import sys
sys.path.append('../..')
import src.Automatic_Topic_Labeling.lable_topic as tl
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.topic_models import TopicModel
```

```python
model = "lda"

vec = Vectorizer.load('vectorizer/lda_without_bigrams_test.pkl')
tm = TopicModel.load('topic_models/lda/lda_without_bigrams_15articles_test.pkl')
docs = dl.get_articles_by_sources(["spiegel","zeit"],kind = "wobigrams")
pos = None

docs = docs["article_texts"]
labels = tl.get_topic_lables([],docs,pos,model,vec,tm,n_top_words = 15, n_cand_lables = 100,lable_min_df = 15, n_labels = 8)

print("\nTopical labels:")
print("-" * 20)
for i, labels in enumerate(labels):
    print(u"Topic {}: {}\n".format(
        i,
        ', '.join(map(lambda l: ' '.join(l), labels))
    ))
```
