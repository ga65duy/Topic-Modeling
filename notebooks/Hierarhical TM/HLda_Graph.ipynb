```python
import sys
sys.path.append('../..')
import src.data.data_loader as dl
from src.features.vectorizer import Vectorizer
from src.models.sampler import HierarchicalLDA
```

### HLda for german editorial articles

```python
data = dl.get_articles_by_type("german","editorial")
texts = data["article_texts"]
```

```python
#vektor aus texten
vec = Vectorizer("tf", texts)
#für jedes dokumnet ein vektor -> bag of words
document_token_matrix = vec.get_document_token_matrix(texts)
vocabX = vec.get_id2token_mapping()
```

```python
vocab2 = vec.get_token_list()
vocab = vec.token2id_mapping
```

```python
doc_converted = []
for text in texts:
    doc = text.split()
    doc = [vocab[word.lower()] for word in doc]
    doc_converted.append(doc)
```

```python
n_samples = 100      # no of iterations for the sampler
alpha = 10          # smoothing over level distributions
gamma = 1.0           # CRP smoothing parameter; number of imaginary customers at next, as yet unused table
eta = 0.01             # smoothing over topic-word distributions
num_levels = 3        # the number of levels in the tree
display_topics = 10   # the number of iterations between printing a brief summary of the topics so far
n_words = 5           # the number of most probable words to print for each topic after model estimation
with_weights = True  # whether to print the words with the weights

hlda = HierarchicalLDA(doc_converted, vocab2, alpha=alpha, gamma=gamma, eta=eta, num_levels=num_levels, verbose=True)
hlda.estimate(n_samples, display_topics=display_topics, n_words=n_words, with_weights=with_weights)
```

```python
print(type(hlda.estimate(n_samples, display_topics=display_topics, n_words=n_words, with_weights=with_weights)))
```

```python
def graph(): 
    dict_list = hlda.dict_for_tree(5)
    from graphviz import Digraph
    dot = Digraph(comment='Topics')
    t = Digraph ("HLda")
    for l in dict_list:
        dot.node(str(l["topic_id"]),l["topic"].replace(",","\n"), shape = "box")
        if l["parent"] != -1:
            dot.edge(str(l["parent"]),str(l["topic_id"]),constraint='true')
    return dot
```

```python
dot = graph()
```

```python
dot.render("German/Editorial", view = True)
```

### HLda english forum

```python
def param_for_Hlda(texts):
    #vektor aus texten
    vec = Vectorizer("tf", texts)
    #für jedes dokumnet ein vektor -> bag of words
    document_token_matrix = vec.get_document_token_matrix(texts)
    vocabX = vec.get_id2token_mapping()
    vocab2 = vec.get_token_list()
    vocab = vec.token2id_mapping
    doc_converted = []
    for text in texts:
        doc = text.split()
        doc = [vocab[word.lower()] for word in doc]
        doc_converted.append(doc)
    return doc_converted,vocab2
```

```python
data = dl.get_forum_threads_by_language("english","forum")
texts = data["thread_texts"]

doc_converted,vocab2 = param_for_Hlda(texts)
```
